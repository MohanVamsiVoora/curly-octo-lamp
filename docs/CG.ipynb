### 6.<br>
### Krylov Subspaces
Consider the "Improvement in x"
<br>
$x_k = x_{k-1} + \alpha p_{k-1}$
<br>
Therefore, $x_k = x_0 + \sum_{i=0}^{k-1} \alpha_i p_{i}$
<br>
As $x_0 = 0$, we can write $x_k \in span\{p_0,p_1,p_2,...p_{k-1}\}$
<br><br>
Consider "Finding the new direction from residual"<br>
$p_k = r_k + \beta_k p_{k-1}$
<br>Making $r_k$ as subject,
$r_k = p_k - \beta p_{k-1} \implies r_k \in span\{p_k,p_{k-1}\}$ , k>1

**Note** here $\alpha$ 's and $\beta$ 's  are non zero. Hence we can say that from the above two observations, 
<br> $span\{p_0,p_1,...p_{k-1}\} = span\{x_1,x_2,x_3,...x_k \} =  span\{r_0,r_1,...r_{k-1}\}$
<br><br>Considering the "Recursive residual Update equation"
<br>
$r_k = r_{k-1} -\alpha_k A p_{k-1}$
<br><br>Given $r_0 =p_0 = b$,

$r_1 = b - \alpha_{k} A b = span \{b, Ab\} \implies p_1 \in span\{b,Ab\}$
<br>$r_2 = r_1 - \alpha_{k} p_1 \in span \{r_1, Ap_1 \} = span\{b,Ab,A^2b\} $  

Extending the above process we get,
<br>
$r_k \in \{b,Ab,A^2b,...A^{k-1}b\}$

So we can conclude that <br>
$span\{p_0,p_1,...p_{k-1}\} = span\{x_1,x_2,x_3,...x_k \} =  span\{r_0,r_1,...r_{k-1}\}
=span\{b,Ab,A^2b,...A^{k-1}b\} = K_k(A,b)$

### Orthogonality of r's and Conjugacy of p's
We need to prove $ r^T_kr_j = 0$ for $p^T_kAp_j = 0$ for j<k.

Given the initial $r_0,p_0,x_0$ the subsequent $r_k,p_k,x_k$ depend only on the 3 update equations.
Now the choice of  $\alpha$ and $\beta$ is responsible for the  properties among r's and p's.

If we can prove that we arrive at these $\alpha's$ and $\beta's$ by considering the orthogonality among r's and A-orthogonality among p's and if transition from each step only involves bi-implication, then it implies we have proven the necessary 

**Note**: $\leftrightarrow$ has been used to represent bi-implication

Given the update equations with unknown $\alpha$ and $\beta$: we will prove
<br>$$p_k^TAp_j = 0 , r_k^Tr_j = 0 , j<k \ \ \leftrightarrow \ \  \alpha_k= \frac{r^T_{k-1}r_{k-1}}{p_{k-1}^TAp_{k-1}}, \beta = \frac{r_k^Tr_k}{r_{k-1}^Tr_{k-1}}$$

So with the update equations and constraints ...
<br>$$x_k = x_{k-1} + \alpha p_{k-1}$$
<br>$$p_k = r_k + \beta_k p_{k-1}$$
<br>$$r_k = r_{k-1} -\alpha_k A p_{k-1}$$
<br>$$p_k^TAp_j = 0 , r_k^Tr_j = 0 , j<k$$

... we will find $\alpha$ and $\beta$

<br>$$r_k = r_{k-1} -\alpha_k A p_{k-1}$$

$\leftrightarrow r_k^Tr_{k-1} = r^T_{k-1}r_{k-1} - \alpha_kp_{k-1}^TAr_{k-1}$

$\leftrightarrow r_k^Tr_{k-1} = r^T_{k-1}r_{k-1} - \alpha_kp_{k-1}^TAp_{k-1}$, since $p_{k-1} = r_{k-1}-\beta_k{p_{k-2}} \ and\  p_{k-1}^T Ap_{k-2} = 0 $

$\leftrightarrow 0 = r^T_{k-1}r_{k-1} - \alpha_kp_{k-1}^TAp_{k-1}$, since $r_k^Tr_{k-1} = 0$

$\leftrightarrow  \alpha_k= \frac{r^T_{k-1}r_{k-1}}{p_{k-1}^TAp_{k-1}} $

Now with the direction update equation

$p_k = r_k + \beta_k p_{k-1}$ $\ \leftrightarrow \ $$A^Tp_k^T = A^Tr_k^T + \beta_k A^Tp_{k-1}^T$$\ \leftrightarrow \ $$p_{k-1}Ap_k^T = p_{k-1}Ar_k^T + \beta_k p_{k-1}Ap_{k-1}^T, \qquad A=A^T$

$\leftrightarrow \ $$ 0 = p_{k-1}Ar_k^T + \beta_k p_{k-1}Ap_{k-1}^T$

$ \ \leftrightarrow \ $$\beta_k = -\frac{p_{k-1}Ar_k^T}{p_{k-1}Ap_{k-1}^T}$ $\ \leftrightarrow \ $$\beta_k = -\frac{\alpha_k}{\alpha_k}\frac{p_{k-1}Ar_k^T}{p_{k-1}Ap_{k-1}^T}$, as $p_{k-1}Ar_k^T = r_{k}Ap_{k-1}^T \ and \ r_k = r_{k-1} -\alpha_k A p_{k-1} $

$\ \leftrightarrow \ $$\beta_k = -\frac{r_k\alpha_kAp_{k-1}^T}{r_{k-1}^Tr_{k-1}}$$\ \leftrightarrow \ $$\beta_k = \frac{r_{k}{(r_k-r_{k-1})}^T}{r_{k-1}^Tr_{k-1}}$$\ \leftrightarrow \ $$\beta_k = \frac{r_{k}^T{r_k}}{r_{k-1}^Tr_{k-1}}$

Thus we have given the update equations with $\alpha$ and $\beta$ as unknowns,
<br>
<br>$$p_k^TAp_j = 0 , r_k^Tr_j = 0 , j<k \ \ \leftrightarrow \ \  \alpha_k= \frac{r^T_{k-1}r_{k-1}}{p_{k-1}^TAp_{k-1}}, \beta = \frac{r_k^Tr_k}{r_{k-1}^Tr_{k-1}}$$

Therefore we have proven the necessary.

### Property of A-orthogonal vectors, A is an spd
<br>
If $p_0,p_1,p_2,..,p_{n-1}$ are n non zero A-orthogonal vectors,then they are linearly independent.
<br>
$\mathbf{Proof \  by \ contradiction:}$
<br><br>
Let $p_{k} = \sum_{i \neq k }\delta_i p_i \implies
 p_{k}^TAp_{k} = \sum_{i \neq k }\delta_i p_i^TAp_{k} = 0$, but A is symmetric positive definite (spd). Hence $p_{k} $ can not be expressed as a weighted  sum of the remaining n-1 vectors.

**Corollary:**
<br>
If A is nxn spd matrix and $p_i$ 's are A -orthogonal, we cannot find more than n such p vectors as all these p vectors need to be linearly independent.
<br>
Therefore, the number of iterations can not be more than n as **we can not have more than n**  "A -orthogonal p vectors".

Thus we can write $x^* = x_ 0 + \sum_{i=0}^{n-1}\alpha_ip_i$ = $\sum_{i=0}^{n-1}\alpha_ip_i, \qquad Ax^* = b $ <br> <br>and $ x_k =\sum_{i=0}^{k-1}\alpha_ip_i$ $ \implies e_k =\left(\sum_{j = k }^{j = n-1} \alpha_j p_j\right)$

### Minimisation of the convex function
$Let \ \phi(x) = \frac{1}{2} x^TAx - x^Tb $<br>
$\|e_n\|_A^2 = (x^*-x_n)^TA(x^*-x_n) = x_n^TAx_n - 2x_n^Tb + x^*b = 2 \phi(x_n) + x^*b , \qquad Ax^* =b $

$\|e_k\|_A^2 = (e_k)^TA(e_n) = \left(\sum_{j = k }^{j = n-1} \alpha_j p_j\right)^TA\left(\sum_{j = k }^{j = n-1} \alpha_j p_j\right)$

From the A -orthogonality of p vectors we can write, 
<br>
$\|e_k\|_A^2 = \sum_{j=k}^{j=n-1} \alpha^2_j p_j^TAp_j$<br>The above error-A-norm only has terms associated with the directions $\{p_k,p_{k+1},...p_{n-1}\}$ that have not been traversed so far.


We can see that $\|e_0\|_A^2 =\alpha^2_0p_0^TAp_0 + \alpha^2_1p_1^TAp_1 + \alpha^2_2p_2^TAp_2 + ... \alpha^2_{n-1}p_{n-1}^TAp_{n-1}$.<br>
Now in each iteration, one term gets removed.<br>
After the kth interation we will be at $x_k$ in the span  of $\{p_0,p_1,p_2,...p_{k-1}\}= K_k(A,b)$. 

$\{p_{k},p_{k+1},...p_{n-1}\}$ are linearly independent with the previous p's.So in what ever way we move inside  $span \{p_0,p_1,p_2,...p_{k-1}\}= K_k(A,b)$ we cannot remove the terms associated with  $p_j,j\geq k$

If we take any other point  $x_k$ in $K_k\left(A,b\right)$, $\|e_k\|_A $ should  have the above terms for sure in addition to some other terms 

Because of the conjugacy among  directions or linear independance , $\|e_k\|_A^2 = \sum_{j=k}^{j=n-1} \alpha^2_j p_j^TAp_j$ has only the terms associated with those directions which are not in $K_k\left(A,b\right)$, $x_k$ will be the unique point in $K_k\left(A,b\right)$ that minimises $\|e\|_A = 2 \phi(x) +constant$

$\|e_k\|_A^2 = \sum_{j=k}^{j=n-1} \alpha^2_j p_j^TAp_j$<br>
$\|e_{k-1}\|_A^2 = \sum_{j=k-1}^{j=n-1} \alpha^2_j p_j^TAp_j = \alpha_{k-1}^2 p_{k-1}^TAp_{k-1}+\|e_k\|_A^2 $

Thus $\|e_{k-1}\|_A<\|e_{k}\|_A$, as $\alpha_{k-1}^2 p_{k-1}^TAp_{k-1}$ is always positive for a spd A.



### Convergence

Already we have seen that we cannot have more than $n$ A-orthogonal p vectors.Therefore mathematically, we can not have more than n iterations.We will now look at, if convergence is possible in lesser than n steps

 $r_{k}\in span\{b,Ab,A^2b,...A^{k}b\} =span\{Ax^*,A^2x^*,A^3x^*,...A^{k+1}x^*\}$

Also we have $r_k = b -Ax_k = Ax^* - A x_k = Ae_k$

Therefore, $Ae_k \in span\{Ax^*,A^2x^*,A^3x^*,...A^{k+1}x^*\} \implies e_k \in span\{x^*,Ax^*,A^2x^*,...A^{k}x^*\}$

$\implies e_k =  \left( I+ \sum_{i=1}^{i = k} c_i A^i \right)x^* = P_k(A)x^*, \qquad P_k(A)$ is a polynomial of order k, **Polynomial stems from Krylov space**

As CG provides us with an $x_k$ such that $\|e_k\|_A $ is minimised in $K_k(A,b)$.We can just rephrase it to say that CG provides us the perfect polynomial $P_k(A)$ of order k, so that minimised $\|e_k\|_A$ over all possible $P_k(A)$ 

Let $x^* = \sum_{j=0}^{n-1} \delta_j \mathbf{v_j}$, $\mathbf{v_i}'s$ are orthogonal unit eigen vectors of spd A.

Substituting the above we get, $\|e_k\|_A^2 = e_k^TAe_k = \sum_j \delta_j^2 P_k^2(\lambda_j) \lambda_j $

Here, P(x) is kth order polynomial in x with P(0)=1.And CG gaurantees the best fit such that  $\sum_{j=1}^{j=n} \delta_j^2 P^2(\lambda_j) \lambda_j$ minimises

Now if $e_k$ were to be zero even before k = n, then

1.
**If some eigen values of A are equal, say m of the n $\lambda_j 's$   are equal.**

$\|e_k\|_A^2  = \sum_j ^{j = n-m}\delta_j^2 P_k^2(\lambda_j) \lambda_j $

Therefore when k = n-m, we can come up with a polynomial P of n-m degree and P(0) = 1, such that $\lambda_1,..\lambda_{n-m}$ are roots of P 

$P(x) = (1-x/\lambda_1)(1-x/\lambda_2)...(1-x/\lambda_{n-m})$, satisfies the above constraints

As CG gives us the best polynomial possible, we are sure that  this is the one.

Therefore,$ e_k = 0$ for  k = n-m

**2. Special case**

If we are fortunate enough we could be solving for  $ x^* = \sum_{i} \delta_i \mathbf{v_i}$ which could be arrived even before traversing all n-m directions.
<br>
This happens when $ \delta_i = 0 $, for some $ i\leq n $.
<br>
In general if we have good starting point, then we can reach $x^*$ with lesser number of iterations.

Therefore we can have $\|e_k\|_A = 0  \ for\  k\leq n$

In practice, however we will have rounding errors and depending on the conditioning of A, it may take more than n steps
